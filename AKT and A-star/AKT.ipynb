```python
import copy
from heapq import heappush, heappop # Fixed: heppush -> heappush

# giả sử puzzle (n=3)
n=3

#4 vị trí dịch chuyển tương ứng bottom, left, top, right
rows=[1,0,-1,0]
cols=[0,-1,0,1]

# tạo một lớp hàng đợi
class priorityQueue:
  def __init__(self):
    self.heap=[]
  # Inserting a new key 'key'
  def push(self, key):
    heappush(self.heap, key)
  #funct to remove the element that is min from the Priority Queue
  def pop(self):
    return heappop(self.heap)
  def empty(self):
    if not self.heap:
      return True
    else:
      return False
#structure of the node
class nodes:
  def __init__(self, parent, mats, empty_tile_posi, costs, levels):
    self.parent=parent

    #Useful for Storing the matrix
    self.mats=mats
    self.empty_tile_posi=empty_tile_posi
    self.costs=costs
    self.level=levels
  def __lt__(self, nxt):
    return self.costs<nxt.costs

def calculateCost(mats, final) -> int:

  count=0
  for i in range(n):
    for j in range(n):
      if mats[i][j] != 0 and mats[i][j] != final[i][j]: # Condition checks if it's not the empty tile (0) and misplaced
        count+=1
  return count
def newNodes(mats, empty_tile_posi, new_empty_tile_posi, levels, parent, final) -> nodes:

  # Copying data from the parent matrixes to the present matrixes
  new_mats=copy.deepcopy(mats)

  #moving the tile by 1 position
  x1 = empty_tile_posi[0]
  y1 = empty_tile_posi[1]
  x2 = new_empty_tile_posi[0]
  y2 = new_empty_tile_posi[1]
  new_mats[x1][y1], new_mats[x2][y2] = new_mats[x2][y2],new_mats[x1][y1]

  # setting the no. of misplaced tiles
  costs=calculateCost(new_mats, final) # Fixed: calculateCosts -> calculateCost

  new_nodes=nodes(parent, new_mats, new_empty_tile_posi,costs, levels)
  return new_nodes

#func to print the N by N matrix
def printMatstrix(mats):

  for i in range(n):
    for j in range(n):
      print("%d " % mats[i][j], end="") # Fixed: (mats[i][j],end= "") -> mats[i][j], end="" (added space for better formatting)

    print()
# func to know if (x,y) is a valid or invalid
# matrix coordinates
def isSafe(x,y):

  return x>=0 and x<n and y>=0 and y<n

#Printing the path from the root node to the final node
def printPath(root):

  if root==None:
    return

  printPath(root.parent)
  printMatstrix(root.mats)
  print()

def solve(initial, empty_tile_posi, final):

  pq=priorityQueue()
  #Creating the root node
  costs=calculateCost(initial, final)
  root=nodes(None, initial, empty_tile_posi, costs, 0)
  # Adding root to the list of live nodes
  pq.push(root)

  while not pq.empty():
    minium=pq.pop()
    if minium.costs==0:
      printPath(minium)
      return

    #Generating all feasible children
    for i in range(4): # Fixed: n -> 4, as rows/cols have 4 directions
      new_tile_posi = [minium.empty_tile_posi[0] + rows[i], minium.empty_tile_posi[1] + cols[i]] # Fixed: `new=` typo

      if isSafe(new_tile_posi[0], new_tile_posi[1]):
        #creating a child node
        child=newNodes(minium.mats,minium.empty_tile_posi,new_tile_posi,minium.level+1,minium,final) # Fixed: minium.levels+1 -> minium.level+1 (typo)

        #adding the child to the list of live nodes
        pq.push(child)

# Example usage - moved outside the solve function and corrected indentation.
initial=[[1,2,3],
         [5,6,0],
         [7,8,4]]
final=[[1,2,3],
       [5,8,6],
       [0,7,4]]
empty_tile_posi=[1,2] # This is the position of 0 in the initial state

print("Solving 8-puzzle:")
solve(initial, empty_tile_posi, final)
```
