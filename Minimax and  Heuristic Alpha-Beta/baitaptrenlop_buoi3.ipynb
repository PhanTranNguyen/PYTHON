```python
#Bài 1: Cài đặt thuật toán Minimax cho Tic-Tac-Toe 3x3
import math

# Định nghĩa các hằng số cho người chơi
PLAYER_X = 'X'
PLAYER_O = 'O'
EMPTY = ' '

def print_board(board):
    """In ra bảng Tic-Tac-Toe."""
    for row in board:
        print(" | ".join(row))
        print("---------")

def check_win(board, player):
    """Kiểm tra xem người chơi 'player' có thắng không."""
    # Kiểm tra hàng
    for row in board:
        if all([cell == player for cell in row]):
            return True
    # Kiểm tra cột
    for col in range(3):
        if all([board[row][col] == player for row in range(3)]):
            return True
    # Kiểm tra đường chéo
    if (board[0][0] == player and board[1][1] == player and board[2][2] == player) or \
       (board[0][2] == player and board[1][1] == player and board[2][0] == player):
        return True
    return False

def check_draw(board):
    """Kiểm tra xem trò chơi có hòa không."""
    for row in board:
        for cell in row:
            if cell == EMPTY:
                return False  # Vẫn còn ô trống, chưa hòa
    return True # Không còn ô trống và không ai thắng

def is_game_over(board):
    """Kiểm tra xem trò chơi đã kết thúc (thắng hoặc hòa) chưa."""
    return check_win(board, PLAYER_X) or check_win(board, PLAYER_O) or check_draw(board)

def minimax(board, depth, is_maximizing_player):
    """
    Thuật toán Minimax để tìm nước đi tốt nhất.
    depth: độ sâu hiện tại của cây tìm kiếm (không sử dụng ở đây vì trò chơi nhỏ).
    is_maximizing_player: True nếu là lượt của người chơi tối đa (AI), False nếu là lượt của người chơi tối thiểu.
    """

    # Kiểm tra các trường hợp cơ sở (trò chơi kết thúc)
    if check_win(board, PLAYER_X):
        return -1 # AI (O) thua, nên giá trị thấp
    if check_win(board, PLAYER_O):
        return 1  # AI (O) thắng, nên giá trị cao
    if check_draw(board):
        return 0  # Hòa, giá trị trung bình

    if is_maximizing_player: # Lượt của AI (PLAYER_O)
        best_value = -math.inf
        for r in range(3):
            for c in range(3):
                if board[r][c] == EMPTY:
                    board[r][c] = PLAYER_O
                    value = minimax(board, depth + 1, False)
                    board[r][c] = EMPTY  # Hoàn tác nước đi
                    best_value = max(best_value, value)
        return best_value
    else: # Lượt của người chơi đối thủ (PLAYER_X)
        best_value = math.inf
        for r in range(3):
            for c in range(3):
                if board[r][c] == EMPTY:
                    board[r][c] = PLAYER_X
                    value = minimax(board, depth + 1, True)
                    board[r][c] = EMPTY  # Hoàn tác nước đi
                    best_value = min(best_value, value)
        return best_value

def find_best_move(board):
    """Tìm nước đi tối ưu cho AI (PLAYER_O) sử dụng Minimax."""
    best_value = -math.inf
    best_move = (-1, -1)

    for r in range(3):
        for c in range(3):
            if board[r][c] == EMPTY:
                board[r][c] = PLAYER_O
                move_value = minimax(board, 0, False)
                board[r][c] = EMPTY  # Hoàn tác nước đi

                if move_value > best_value:
                    best_value = move_value
                    best_move = (r, c)
    return best_move

def play_game():
    """Chạy trò chơi Tic-Tac-Toe với AI sử dụng Minimax."""
    board = [
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY]
    ]
    current_player = PLAYER_X # Người chơi X đi trước

    print("\n--- BẮT ĐẦU TRÒ CHƠI TIC-TAC-TOE --- ")
    print_board(board)

    while not is_game_over(board):
        if current_player == PLAYER_X:
            print("Lượt của bạn (X). Nhập hàng và cột (0-2):")
            while True:
                try:
                    row = int(input("Hàng: "))
                    col = int(input("Cột: "))
                    if 0 <= row <= 2 and 0 <= col <= 2 and board[row][col] == EMPTY:
                        board[row][col] = PLAYER_X
                        break
                    else:
                        print("Nước đi không hợp lệ. Vui lòng thử lại.")
                except ValueError:
                    print("Đầu vào không hợp lệ. Vui lòng nhập số.")
        else: # Lượt của AI (PLAYER_O)
            print("Lượt của máy (O)...")
            row, col = find_best_move(board)
            board[row][col] = PLAYER_O
            print(f"Máy đặt 'O' vào ô ({row}, {col})")

        print_board(board)

        if check_win(board, PLAYER_X):
            print("Chúc mừng! Bạn (X) đã thắng!")
            return
        elif check_win(board, PLAYER_O):
            print("Máy (O) đã thắng! Bạn thua cuộc.")
            return
        elif check_draw(board):
            print("Trò chơi hòa!")
            return

        # Đổi lượt người chơi
        current_player = PLAYER_O if current_player == PLAYER_X else PLAYER_X

# Chạy trò chơi
play_game()
```

```python
import math

# Định nghĩa các hằng số cho người chơi
PLAYER_X = 'X'
PLAYER_O = 'O'
EMPTY = ' '

def print_board(board):
    """In ra bảng Tic-Tac-Toe."""
    for row in board:
        print(" | ".join(row))
        print("---------")

def check_win(board, player):
    """Kiểm tra xem người chơi 'player' có thắng không."""
    # Kiểm tra hàng
    for row in board:
        if all([cell == player for cell in row]):
            return True
    # Kiểm tra cột
    for col in range(3):
        if all([board[row][col] == player for row in range(3)]):
            return True
    # Kiểm tra đường chéo
    if (board[0][0] == player and board[1][1] == player and board[2][2] == player) or \
       (board[0][2] == player and board[1][1] == player and board[2][0] == player):
        return True
    return False

def check_draw(board):
    """Kiểm tra xem trò chơi có hòa không."""
    for row in board:
        for cell in row:
            if cell == EMPTY:
                return False  # Vẫn còn ô trống, chưa hòa
    return True # Không còn ô trống và không ai thắng

def is_game_over(board):
    """Kiểm tra xem trò chơi đã kết thúc (thắng hoặc hòa) chưa."""
    return check_win(board, PLAYER_X) or check_win(board, PLAYER_O) or check_draw(board)

def minimax(board, depth, is_maximizing_player):
    """
    Thuật toán Minimax để tìm nước đi tốt nhất.
    depth: độ sâu hiện tại của cây tìm kiếm (không sử dụng ở đây vì trò chơi nhỏ).
    is_maximizing_player: True nếu là lượt của người chơi tối đa (AI), False nếu là lượt của người chơi tối thiểu.
    """

    # Kiểm tra các trường hợp cơ sở (trò chơi kết thúc)
    if check_win(board, PLAYER_X):
        return -1 # AI (O) thua, nên giá trị thấp
    if check_win(board, PLAYER_O):
        return 1  # AI (O) thắng, nên giá trị cao
    if check_draw(board):
        return 0  # Hòa, giá trị trung bình

    if is_maximizing_player: # Lượt của AI (PLAYER_O)
        best_value = -math.inf
        for r in range(3):
            for c in range(3):
                if board[r][c] == EMPTY:
                    board[r][c] = PLAYER_O
                    value = minimax(board, depth + 1, False)
                    board[r][c] = EMPTY  # Hoàn tác nước đi
                    best_value = max(best_value, value)
        return best_value
    else: # Lượt của người chơi đối thủ (PLAYER_X)
        best_value = math.inf
        for r in range(3):
            for c in range(3):
                if board[r][c] == EMPTY:
                    board[r][c] = PLAYER_X
                    value = minimax(board, depth + 1, True)
                    board[r][c] = EMPTY  # Hoàn tác nước đi
                    best_value = min(best_value, value)
        return best_value

def find_best_move(board):
    """Tìm nước đi tối ưu cho AI (PLAYER_O) sử dụng Minimax."""
    best_value = -math.inf
    best_move = (-1, -1)

    for r in range(3):
        for c in range(3):
            if board[r][c] == EMPTY:
                board[r][c] = PLAYER_O
                move_value = minimax(board, 0, False)
                board[r][c] = EMPTY  # Hoàn tác nước đi

                if move_value > best_value:
                    best_value = move_value
                    best_move = (r, c)
    return best_move

def play_game():
    """Chạy trò chơi Tic-Tac-Toe với AI sử dụng Minimax."""
    board = [
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY],
        [EMPTY, EMPTY, EMPTY]
    ]
    current_player = PLAYER_X # Người chơi X đi trước

    print("\n--- BẮT ĐẦU TRÒ CHƠI TIC-TAC-TOE --- ")
    print_board(board)

    while not is_game_over(board):
        if current_player == PLAYER_X:
            print("Lượt của bạn (X). Nhập hàng và cột (0-2):")
            while True:
                try:
                    row = int(input("Hàng: "))
                    col = int(input("Cột: "))
                    if 0 <= row <= 2 and 0 <= col <= 2 and board[row][col] == EMPTY:
                        board[row][col] = PLAYER_X
                        break
                    else:
                        print("Nước đi không hợp lệ. Vui lòng thử lại.")
                except ValueError:
                    print("Đầu vào không hợp lệ. Vui lòng nhập số.")
        else: # Lượt của AI (PLAYER_O)
            print("Lượt của máy (O)...")
            row, col = find_best_move(board)
            board[row][col] = PLAYER_O
            print(f"Máy đặt 'O' vào ô ({row}, {col})")

        print_board(board)

        if check_win(board, PLAYER_X):
            print("Chúc mừng! Bạn (X) đã thắng!")
            return
        elif check_win(board, PLAYER_O):
            print("Máy (O) đã thắng! Bạn thua cuộc.")
            return
        elif check_draw(board):
            print("Trò chơi hòa!")
            return

        # Đổi lượt người chơi
        current_player = PLAYER_O if current_player == PLAYER_X else PLAYER_X

# Chạy trò chơi
play_game()
```
