```python
#bai 3
import math

# Kiểm tra thắng
def check_winner(board, player, win_len):
    n = len(board)

    # kiểm tra hàng
    for i in range(n):
        for j in range(n - win_len + 1):
            if all(board[i][j + k] == player for k in range(win_len)):
                return True

    # kiểm tra cột
    for j in range(n):
        for i in range(n - win_len + 1):
            if all(board[i + k][j] == player for k in range(win_len)):
                return True

    # đường chéo chính
    for i in range(n - win_len + 1):
        for j in range(n - win_len + 1):
            if all(board[i + k][j + k] == player for k in range(win_len)):
                return True

    # đường chéo phụ
    for i in range(n - win_len + 1):
        for j in range(win_len - 1, n):
            if all(board[i + k][j - k] == player for k in range(win_len)):
                return True

    return False


# kiểm tra còn nước đi không
def is_full(board):
    return all(cell != '.' for row in board for cell in row)


# hàm đánh giá (heuristic đơn giản)
def evaluate(board, win_len):
    if check_winner(board, 'X', win_len):
        return 10
    elif check_winner(board, 'O', win_len):
        return -10
    return 0


# Minimax có depth limit
def minimax(board, depth, is_max, win_len, max_depth):
    score = evaluate(board, win_len)

    # nếu có kết quả
    if score == 10 or score == -10:
        return score

    if is_full(board) or depth == max_depth:
        return 0

    n = len(board)

    if is_max:
        best = -math.inf

        for i in range(n):
            for j in range(n):
                if board[i][j] == '.':
                    board[i][j] = 'X'
                    val = minimax(board, depth + 1, False, win_len, max_depth)
                    board[i][j] = '.'
                    best = max(best, val)

        return best

    else:
        best = math.inf

        for i in range(n):
            for j in range(n):
                if board[i][j] == '.':
                    board[i][j] = 'O'
                    val = minimax(board, depth + 1, True, win_len, max_depth)
                    board[i][j] = '.'
                    best = min(best, val)

        return best


# tìm nước đi tốt nhất
def find_best_move(board, win_len, max_depth=3):
    n = len(board)
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(n):
        for j in range(n):
            if board[i][j] == '.':
                board[i][j] = 'X'
                move_val = minimax(board, 0, False, win_len, max_depth)
                board[i][j] = '.'

                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val

    return best_move
```

```python
#bai 4
import math

# kiểm tra thắng
def check_winner(board, player, win_len):
    n = len(board)

    # hàng
    for i in range(n):
        for j in range(n - win_len + 1):
            if all(board[i][j + k] == player for k in range(win_len)):
                return True

    # cột
    for j in range(n):
        for i in range(n - win_len + 1):
            if all(board[i + k][j] == player for k in range(win_len)):
                return True

    # chéo chính
    for i in range(n - win_len + 1):
        for j in range(n - win_len + 1):
            if all(board[i + k][j + k] == player for k in range(win_len)):
                return True

    # chéo phụ
    for i in range(n - win_len + 1):
        for j in range(win_len - 1, n):
            if all(board[i + k][j - k] == player for k in range(win_len)):
                return True

    return False


def is_full(board):
    return all(cell != '.' for row in board for cell in row)


# heuristic đơn giản
def evaluate(board, win_len):
    if check_winner(board, 'X', win_len):
        return 100
    elif check_winner(board, 'O', win_len):
        return -100
    return 0


# Alpha-Beta
def alphabeta(board, depth, alpha, beta, is_max, win_len, max_depth):
    score = evaluate(board, win_len)

    # terminal
    if score == 100 or score == -100:
        return score

    if is_full(board) or depth == max_depth:
        return 0

    n = len(board)

    if is_max:
        best = -math.inf

        for i in range(n):
            for j in range(n):
                if board[i][j] == '.':
                    board[i][j] = 'X'

                    val = alphabeta(board, depth + 1, alpha, beta, False, win_len, max_depth)

                    board[i][j] = '.'
                    best = max(best, val)

                    alpha = max(alpha, best)

                    # cắt tỉa
                    if beta <= alpha:
                        return best

        return best

    else:
        best = math.inf

        for i in range(n):
            for j in range(n):
                if board[i][j] == '.':
                    board[i][j] = 'O'

                    val = alphabeta(board, depth + 1, alpha, beta, True, win_len, max_depth)

                    board[i][j] = '.'
                    best = min(best, val)

                    beta = min(beta, best)

                    # cắt tỉa
                    if beta <= alpha:
                        return best

        return best


# tìm nước đi tốt nhất
def find_best_move(board, win_len, max_depth=3):
    n = len(board)
    best_val = -math.inf
    best_move = (-1, -1)

    for i in range(n):
        for j in range(n):
            if board[i][j] == '.':
                board[i][j] = 'X'

                move_val = alphabeta(
                    board,
                    0,
                    -math.inf,
                    math.inf,
                    False,
                    win_len,
                    max_depth
                )

                board[i][j] = '.'

                if move_val > best_val:
                    best_val = move_val
                    best_move = (i, j)

    return best_move
```
